```python
from statsmodels.tsa.stattools import adfuller
# Run Augmented Dickey Fuller Test
result = adfuller(data['s1a1wd'])
# Print p-value
print(result[1])
```

Running the Dickey Fuller test on the 1st principal component, I got a p-value of 1.424376679120803e-14 which is very small number (much smaller than 0.05). Thus, I will reject the Null Hypothesis and say the data is stationary.

```python
#https://analyticsindiamag.com/quick-way-to-find-p-d-and-q-values-for-arima/
from statsmodels.tsa.arima.model import ARIMA
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(data['s1a1wd'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
  print('\t%s: %.3f' % (key, value))
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(data.s1a1wd.diff().dropna())
```

Here we can see that the first 3 lag are significantly out of the limit
and the fourth one is near significant limit  we can select the order of the p as 4. 

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(data.s1a1wd.diff().dropna())
```

Here we can see that 2 of the lags are out of the significance limit so we will assume optimal value of our q (MA) is 2. 

```python
model = ARIMA(data, order=(4,1,1))
model_fit = model.fit()
```

```python

# summary of fit model
print(model_fit.summary())
```

```python
from pandas import DataFrame
from matplotlib import pyplot
```

```python
# line plot of residuals
residuals = DataFrame(model_fit.resid)
residuals.plot()
pyplot.show()
```

```python
# density plot of residuals
residuals.plot(kind='kde')
pyplot.show()
```

```python
# split into train and test sets
X = data.values
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
```

```python
# walk-forward validation
for t in range(len(test)):
 model = ARIMA(history, order=(4,1,2))
 model_fit = model.fit()
 output = model_fit.forecast()
 yhat = output[0]
 predictions.append(yhat)
 obs = test[t]
 history.append(obs)
 print('predicted=%f, expected=%f' % (yhat, obs))
```

```python
# evaluate forecasts
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse = sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % rmse)
```

In ARIMA modeling, RMSE stands for Root Mean Square Error, which is a commonly used measure of the difference between the predicted values and the actual values of a time series.

An RMSE value of 0.007 means that the average difference between the predicted values and the actual values of the time series is 0.007 units. This indicates that the ARIMA model is performing well in predicting the values of the time series with a high degree of accuracy.

However, it is important to note that the interpretation of RMSE values depends on the specific context and the scale of the data being analyzed. In some cases, an RMSE value of 0.007 may be considered high or low depending on the specific application and domain of the problem.

```python

# plot forecasts against actual outcomes
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```

Here we can see that the values are pretty close to the real values. 

If you are finding that ARIMA is not providing accurate results, there are several other models you can consider. Some of the commonly used models for time series analysis are:

Seasonal ARIMA (SARIMA): SARIMA is an extension of ARIMA that includes seasonal components to capture periodic patterns in the data.

Exponential Smoothing (ETS): ETS models are a family of models that include simple exponential smoothing, Holt's linear trend method, and Holt-Winters method that can handle both trend and seasonality.

Vector Autoregression (VAR): VAR models are used to analyze the interdependencies among multiple time series variables.

Long Short-Term Memory (LSTM) Networks: LSTMs are a type of recurrent neural network that can capture complex nonlinear relationships in time series data.

Prophet: Prophet is a forecasting model developed by Facebook that combines an additive model with seasonality, trends, and holidays.

It is important to note that different models have different assumptions and requirements, and it is important to select the appropriate model based on the characteristics of your data
