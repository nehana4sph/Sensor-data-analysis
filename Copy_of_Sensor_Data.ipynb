```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/whole_dataset.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
df = df.drop_duplicates(subset=['Time'],keep='first')
```

```python
df.head()
```

```python
df.rename(columns = {'s1a1wod':'s1a1wd','s1a2wod': 's1a2wd','s1a1wd':'s1a1wod','s1a2wd': 's1a2wod'}, inplace = True)
```

```python
df.describe()
```

```python
df_speed1_wod_left = df[['Time','s1a1wod']]
```

```python
df_speed1_wd_left = df[['Time','s1a1wd']]
```

```python
df_speed1_wd_left.describe()
```

```python
df_speed1_wod_left.describe()
```

```python
data = df_speed1_wod_left
data.describe()
```

```python
data['Time'] = pd.to_datetime(data['Time'], unit='s')
```

```python
data.head()
```

```python
data.info()
```

```python
# Visualize data using seaborn
sns.set(rc={'figure.figsize':(12,8)})
sns.lineplot(x=data['Time'], y=data['s1a1wod'])
plt.legend(['Time Plot'])
plt.title('Time Plot ')
plt.show()
```

```python
data = data.set_index('Time')
```

```python
# Resample the entire dataset by daily average
rollmean = data.resample(rule='D').mean()
rollstd = data.resample(rule='D').std()
```

```python
print(rollmean.head())
print(rollstd.head())
```

```python
# Plot time series for each sensor with its mean and standard deviation
plt.figure(figsize=(18,3))
plt.plot(data['s1a1wod'], color='blue', label='Original')
plt.plot(rollmean['s1a1wod'], color='red', label='Rolling Mean')
plt.plot(rollstd['s1a1wod'], color='black', label='Rolling Std' )
plt.legend(loc='best')
plt.title('Speed 1 Acceleration Left Sensor')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
# Run Augmented Dickey Fuller Test
result = adfuller(data['s1a1wod'])
# Print p-value
print(result[1])
```

Running the Dickey Fuller test on the 1st principal component, I got a p-value of 1.424376679120803e-14 which is very small number (much smaller than 0.05). Thus, I will reject the Null Hypothesis and say the data is stationary.
